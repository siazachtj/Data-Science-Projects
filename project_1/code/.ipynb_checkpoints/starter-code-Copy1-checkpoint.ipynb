## Background

According to the [Meteorological Services Singapore](http://www.weather.gov.sg/climate-climate-of-singapore/#:~:text=Singapore%20is%20situated%20near%20the,month%2Dto%2Dmonth%20variation.), Singapore has typical tropical climate with adundant rainfall, high and uniform temperatures and high humidity all year round, since its situated near the equator. There are many factors that help us understand the climate of a country and in this project we are going to look into a few, especially rainfall.

Singapore’s climate is characterised by two main monsoon seasons separated by inter-monsoonal periods.  The **Northeast Monsoon** occurs from December to early March, and the **Southwest Monsoon** from June to September.

The major weather systems affecting Singapore that can lead to heavy rainfall are:

-Monsoon surges, or strong wind episodes in the Northeast Monsoon flow bringing about major rainfall events;

-Sumatra squalls, an organised line of thunderstorms travelling eastward across Singapore, having developed over the island of Sumatra or Straits of Malacca west of us;

-Afternoon and evening thunderstorms caused by strong surface heating and by the sea breeze circulation that develops in the afternoon.

Singapore’s climate station has been located at several different sites in the past 140 years. The station had been decommissioned at various points in the past due to changes to local land use in the site’s vicinity, and had to be relocated. Since 1984, the climate station has been located at **Changi**.

There are other metrics of climate such as temperature, humidity, sun shine duration, wind speed, cloud cover etc. All the dataset used in the project comes from [data.gov.sg](data.gov.sg), as recorded at the Changi climate station

```python
#functions
#function for handling index before join
def index_handler(df):
    dict_ind = {}
    for idex in df.index:
        if idex[5] == '0':
            dict_ind[idex] = idex[:5]+ idex[6:8]
        else:
            dict_ind[idex] = idex
    return df.rename(index = dict_ind, inplace=True)

#outliers are anything outside the Quartile 1-(1.5*InterQuartileRange and Quartile3+1.5*InterQuartileRange
def outliers(x):
    Q1 = df[x].quantile(.25)
    Q3 = df[x].quantile(.75)
    IQR = Q3 - Q1
    l_bound = Q1 -(1.5 * IQR)
    u_bound = Q3 +(1.5 * IQR)
    print(x, 'outlier values')
    print(df[(df[x] < l_bound) | (df[x] > u_bound)])
    print('====================================================')

    
#function for showing dataframe basic info
def df_explorer(x):
    print(f'DataFrame : {x}.head')
    x.head()
    print(f'DataFrame : {x}.shape')
    x.shape
    print(f'DataFrame : {x}.info')
    x.info()
    print(f'DataFrame : {x}.describe')
    x.describe()
    print(f'DataFrame : {x}.all null values')
    x.isna().sum()
    print("\n")

#function to break year month into seperate columns
def month_year_creator(row):
    year = int(row['year_month'][:4])
    month = int(row['year_month'][5:])
    return year,month
    
```

### mean function

```python
def mean_fun(numbers):
    num_sum = 0
    for i in numbers:
        num_sum +=i
    return num_sum/len(numbers)
mean_fun([9,5,12,8])
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

### standard deviation function

```python
def standard_deviation(numbers):
    counter = 0
    mean = mean_fun(numbers)
    for number in numbers:
        counter +=(number - mean) ** 2
    return (counter/len(numbers))**0.5
standard_deviation([2,4,6,8])
```

### Cleaning RTA_monthly_df
To clean RTA_monthly_df the following was done:
1. column in RTA_monthly_df is formatted as : "2009 Jan", " 2010 March" etc needed to be reformatted to be more numeric. This was achieved by:
    - Generating a list of tuples corresponding to each month e.g. [(1,Jan)..(11,Nov)]
    - renaming **ALL** columns in RTA_monthly_df based on the corresponding value in the tuple e.g. "2022 Jan" becomes "2022-1"
    
    
2. The dataframe was then transposed to match the other three dataframes which had month as a column.(later turned into an index for joining)

3. The focal point was decided to be the Total Casualties Fatalities and Total Casualities Injuries, as these two are summarised versions of all Road Traffic Accidents that occur. 

3. Confusing names were renamed e.g. "Total Casualties Fatalities" to "Total_Fatalities" for readability and easy reference



```python
#programmatically split list into year and month and used dict comp to turn into "mapper" passed into rename function
#column in RTA_monthly_df is formatted as : "Jan 2009" etc. Split this into a list for each column name.
#using 2019 as a base case as it has all months possible (1-12). 
#enumerate generates an index of sorts e.g. (0,Jan).
enum_list = list(enumerate([cols.split(" ")[1] 
                            #for each column in RTA_monthly_df
                            for cols in RTA_monthly_df.columns
                            #after splitting if the Year Month, if Year matches 2019, reverse list (it was originally ordered from Dec to Jan)
                            if cols.split(" ")[0] == '2019'][::-1]))


#using our newly mapped number to month list, map using dictionary comprehension to rename the columns.
RTA_monthly_df.rename({cols: f"{cols.split(' ')[0]}-{i+1}" 
                       #for values in tuple
                       for i,v in enum_list 
                       #for every column in RTA_monthly_df if the month name matches, use corresponding tuple value 
                       for cols in RTA_monthly_df.columns if v == cols.split(" ")[1]},axis=1,inplace=True)


#set the index to be the month and transpose the dataframe
x = RTA_monthly_df.set_index('Data Series').T

#rename the column containing months to be month
x.columns.name = 'month'

#limit search to only total casualities fatalities and injured, cleaned column names with "_" as those are most relevant information
#used temp variables for cleaning
x2 = x[['Total Casualties Fatalities','Total Casualties Injured']]
x2.rename(columns = {i : "_".join(i.split(" ")) for i in x2.columns}, inplace=True)

#changed from str to int due to inconsistencies in the data
x2['Total_Casualties_Fatalities'] = x2['Total_Casualties_Fatalities'].astype("int64")
x2['Total_Casualties_Injured'] = x2['Total_Casualties_Injured'].astype("int64")

#created a combined column
x2['Total Casualties'] = x2['Total_Casualties_Injured'] + x2['Total_Casualties_Fatalities']

#renamed RTA dataframe to RTA_final_df from temp variables
RTA_final_df = x2
RTA_final_df.rename(columns = {'Total_Casualties_Fatalities': 'Total_Fatalities', 'Total_Casualties_Injured' : 'Total_Injuries', 'Total Casualties' : 'Total_Fatalities_and_Injuries'},inplace=True)
df_explorer(RTA_final_df)
```

### Preparing for joining
Used date as the index for joining.

```python
#set index of all to month to prepare for join
rainfall_monthly_volume_df.set_index('month',inplace=True)
rainfall_monthly_days_total_df.set_index('month',inplace=True)
sunshine_monthly_df.set_index('month',inplace=True)
```

```python

# #standardised from 01,02,03 etc to 1,2,3 for easier join
[index_handler(i) for i in [sunshine_monthly_df,rainfall_monthly_days_total_df,rainfall_monthly_volume_df]]
#joined all dataframes
df = RTA_final_df.join([sunshine_monthly_df,rainfall_monthly_days_total_df,rainfall_monthly_volume_df])
# removed na values 
df.dropna(inplace=True)
df.reset_index(inplace=True)
df.rename(columns = {"index":"year_month"},inplace=True)
df.rename(columns = {"year,month":"year_month"},inplace=True)
df_explorer(df)
```

```python
#seperating year month in df
df[['year','month']] = df.apply(month_year_creator, axis=1, result_type='expand')
```

```python
#saved df to csv file for easy reference
if not os.path.exists('../data/output'):
    os.makedirs('../data/output')
df.to_csv('../data/output/cleaned_df.csv', index=False)
```

### Preparation for future investigation.
Below is a loop that displays the correlation and flags any strongly correlated numbers. The intention is to flag the most strongly correlated variables. It is not perfect as it flags the correlation between total fatalities and injuries to the combined value. In future iterations, perhaps the combined value could be removed or the loop can be improved.

```python
for i in df2['year'].unique():
    print(i, 'strongly correlated factors \n')
    zz = df2[df2['year'] == i].corr().values
    rows, cols = zz.shape
    col_list = list(df2[df2['year'] == i].corr().columns)
    for a in range(cols):
        for j in range(a+1, cols):
            if zz[a,j] > 0.5:
                print(f"Factor: {col_list[a]}, and Factor: {col_list[j]}, ' are strongly correlated', {zz[a,j]}")
            elif zz[a,j] < -0.5:
                print(f"Factor: {col_list[a]}, and Factor: {col_list[j]}, ' are strongly correlated', {zz[a,j]}")
    print('==============================================================================================================\n')
```
