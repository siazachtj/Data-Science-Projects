```python
import pandas as pd
```

```python
main_data = pd.read_csv('gamesdata.csv')
main_data
```

```python
first_ten = pd.read_csv('first_ten_k_games.csv')
ten2twenty = pd.read_csv('10to20_k_games.csv')
twenty2thirty = pd.read_csv('twenty_k_games.csv')
thirty2fourty = pd.read_csv('30to40_k_games.csv')
fourty2fifty = pd.read_csv('40to50_k_games.csv')
```

```python
game_data2 = pd.concat([first_ten,ten2twenty,twenty2thirty,thirty2fourty,fourty2fifty], axis=0, ignore_index=True)
game_data2
```

```python
master_df = main_data[:50000]
master_df
```

```python
master_df = pd.concat([master_df,game_data2], axis=1)
master_df = master_df.rename(columns = {'Name':'title','Link':'link','Release Date':'r_date','Price':'price','tags':'tagged'})
master_df
```

```python
master_df.to_csv('master_df3.csv', index=False)
```
