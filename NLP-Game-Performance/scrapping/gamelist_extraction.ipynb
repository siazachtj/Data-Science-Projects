# About This Notebook

This notebook contains code that we use to scrap a list of games from the Steam webpage. 


```python
# Importing libraries

from selenium import webdriver
from bs4 import BeautifulSoup
import time
import requests
import pandas as pd
import json
```

```python
%%time

url = "https://store.steampowered.com/search/?sort_by=Released_DESC&ignore_preferences=1&category1=998&supportedlang=english&ndl=1"

# Start a new Chrome browser instance
driver = webdriver.Chrome()
options = webdriver.ChromeOptions()
options.headless = True


# Navigate to the URL
driver.get(url)

x=0

while x <= 1200: 
    # Scroll down to load the entire page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # Wait for the page to load
    time.sleep(5)
    
    x = x+1

   
    
    # Parse the HTML content
soup = BeautifulSoup(driver.page_source, "html.parser")

# Find all the games on the page
games = soup.find_all("a", class_="search_result_row")

# Initialize an empty list to store the data
data = []

# Loop through each game
for game in games:
    # Find the name, link, price, and description of the game
    name = game.find("span", class_="title").text
    link = game.get("href")
    release = game.find('div', class_='col search_released responsive_secondrow').text
    price = game.find("div", class_="col search_price_discount_combined responsive_secondrow").text.strip()
    
        
    # Append the data to the list 
    data.append([name, link, release, price])

    
    
    
# Close the browser
driver.quit()
```

```python
# Create a Pandas dataframe from the list
df = pd.DataFrame(data, columns=["Name", "Link", 'Release Date', "Price"])

# Print the dataframe
df
```

```python
df.to_csv('dataset/gamesdata.csv', index=False)
```
